## Logistic regression
- it works when data is linearly seprable

### Problem with logistic regression 
- Complex data
- Classes overlap
- Decision boundry is not straight


### We use SVM because of problems of logistic regression.

- Linear
- Polynomial
- RBF
- Sigmoid

- support vectors are the nearest points to the decision boundry
- hyperplane class are parallel lines to decision boundry that seprates two classes and maximise the distance
- decision boundary is drawn between two hyperplanes to seprate two classes

y=mx+c

- parameters are the values that are learned from the data
- hyperparameters are values that not at all depends on the data
- hyperparameters are input parameters which will be passed to the model
- for each and evey hyperparameter , input values  will change for all.
