```python
import numpy as np
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
data=pd.read_csv('framingham.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
# Check missing values
print(data.isnull().sum())

# Fill categorical/binary columns with mode
data['education'].fillna(data['education'].mode()[0], inplace=True)
data['BPMeds'].fillna(data['BPMeds'].mode()[0], inplace=True)

# Fill numerical columns with median
data['cigsPerDay'].fillna(data['cigsPerDay'].median(), inplace=True)
data['totChol'].fillna(data['totChol'].median(), inplace=True)
data['BMI'].fillna(data['BMI'].median(), inplace=True)
data['heartRate'].fillna(data['heartRate'].median(), inplace=True)
data['glucose'].fillna(data['glucose'].median(), inplace=True)

# Verify
print(data.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.boxplot(data)

plt.show()
```

```python
def cap_outlier(column):
    Q1=data[column].quantile(0.25)
    Q3=data[column].quantile(0.75)
    IQR=Q3-Q1
    Lower = Q1 - 1.5*IQR
    Upper = Q3 + 1.5*IQR
    data[column]=data[column].clip(Lower,Upper) #fill the outliers with the upper and lower bound values
columns=data.columns
for col in columns:
    cap_outlier(col)
```

```python
x = data.drop('TenYearCHD',axis=1)
y = data['TenYearCHD']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
svm_regressor = SVR(kernel='rbf', degree=3, gamma='scale', coef0=0.0)
```

```python
svm_regressor.fit(x_train, y_train)
```

```python
y_pred=svm_regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y_test,y_pred)
print("MSE:",mse)
```
